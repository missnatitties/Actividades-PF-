```python
#Actividad: Puntaje Reading
import pandas 
import numpy
df = pandas.read_csv (r'C:\Users\ncordova\Documents\Python Fundamentals\clean_students_complete.csv')
df.head()
```

```python
#Crea / Presenta la forma que creas más conveniente para describir cuántos alumnos por escuela tuvieron el mejor puntaje en "reading", y con ello saber qué género es el que predomina en este filtro.
best_students = df[df["reading_score"]==df["reading_score"].max()]
best_students.head()
```

```python
from matplotlib import pyplot as plt
for_pie = best_students[['school_name', 'reading_score']].groupby('school_name').count()
pie_plot = for_pie['reading_score'].plot.pie(figsize = (10,10), legend=True)
print(for_pie)
```

```python
#Qué genero predominó más
for_pie = best_students[['gender', 'reading_score']].groupby('gender').count()
print(for_pie)
pie_plot = for_pie['reading_score'].plot.pie(figsize = (10,10), legend=True)
```

```python
#Crea / Presenta la forma que creas más conveniente para describir cuántos alumnos por escuela tuvieron el peor puntaje en "reading", y con ello saber que genero es el que predomina en este filtro.
worse_students = df[df["reading_score"]==df["reading_score"].min()]
worse_students.head()
```

```python
for_pie = worse_students[['school_name', 'reading_score']].groupby('school_name').count()
pie_plot = for_pie['reading_score'].plot.pie(figsize = (10,10), legend=True)
print(for_pie)
```

```python
#Qué genero predominó más
for_pie = worse_students[['gender', 'reading_score']].groupby('gender').count()
print(for_pie)
pie_plot = for_pie['reading_score'].plot.pie(figsize = (10,10), legend=True)
```
