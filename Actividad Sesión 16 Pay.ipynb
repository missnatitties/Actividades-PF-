```python
#Importar
import pandas 
import numpy
from matplotlib import pyplot as plt
df = pandas.read_csv (r'C:\Users\ncordova\Documents\Python Fundamentals\athlete_events.csv')
```

```python
#Crea y guarda una gráfica de pay del año más reciente de la base de datos.
recent_yr = df[df['Year']==df['Year'].max()]
for_pie = recent_yr[['NOC', 'Medal']].replace('', numpy.nan).groupby('NOC').count().sort_values('Medal', ascending=False).head(15)
```

```python
pie_plot = for_pie['Medal'].plot.pie(figsize = (10,10), legend=True)
```
